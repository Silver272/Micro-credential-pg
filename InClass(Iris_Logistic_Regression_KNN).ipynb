```python
# Load our data from github
df = pd.read_csv('https://raw.githubusercontent.com/Silver272/ML-Data/main/iris.csv')
```

```python
df.head()
```

```python
sns.pairplot(df,hue='Class',palette='coolwarm')
```

```python
df.describe()
```

Standardize data and drop CLASS variable

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.drop('Class',axis=1))
```

```python
scaled_features = scaler.transform(df.drop('Class',axis=1))
```

```python
scaled_features
```

```python
df_feat = pd.DataFrame(scaled_features,columns=df.columns[:-1])
df_feat.head()
# Data is scaled now
```

Splitting data into train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_features,df['Class'], test_size=0.30)
```

Using the KNN model: Here look at Class('Setosa','Vesicolor' or 'Virginica')

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Start value of K at 1 (K=1)
knn = KNeighborsClassifier(n_neighbors=1) 
```

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

Now that we’ve created our model, lets evaluate how well it performs. Recall from our previous lesson the confusion matrix Not bad – lets see if we can improve it.

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

Change the n_neighbors= value to something else … try to see how it impacts your confusion matrix. Search online for a way to determine the K Value and try to apply to your model

```python
error_rate = []

for i in range(1,40):

  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train,y_train)
  pred_i = knn.predict(X_test)
  error_rate.append(np.mean(pred_i != y_test))
```

```python
error_rate
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs K Value')
plt.xlabel('Error Rate')
```

```python
knn = KNeighborsClassifier(n_neighbors=15)
```

```python
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
```

```python
print ('With k = 15')
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

With k = 15 error rate is very low approaching 0.02, this further shows why accuracy is 96%. error rate is inversely proportional to accuracy.

```python
knn = KNeighborsClassifier(n_neighbors=29)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
print ('With k = 29')
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

As previously stated error rate is inversely proportional to accuracy. With k=29 we have an error rate between 0.06 and 0.08 accuracy reduced to 93% because error rate increased.

```python
knn = KNeighborsClassifier(n_neighbors=40)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
print ('With k = 40')
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

Now k=40 which has the highest error rate at above 0.10 error rate we see that accuracy further drop to 89%. This concludes that error rate is inversely related to accuracy in this analysis.


SVM model

```python
X = df.iloc[:,:-1]
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 0)
```

```python
from sklearn.svm import SVC
```

```python
classifier = SVC(kernel = 'linear', random_state = 0)
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test) 
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, cv = 8)
print("Accuracy: {:.2f} %".format(accuracies.std()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
